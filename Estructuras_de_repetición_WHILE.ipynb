##ESTRUCTURA DE REPETICIÓN MIENTRAS
Es una estrcutura de repetición boleana empelada cuando no se sabe cuantas veces se debe repetir la condición comprende de
1. Una condición inicial
2. Un centinela que evalua si se cumple la condición final
3. Un modificador de condición inicial

A diferencia de la estructura para, eñ modificador puede ser un contador, un acumulador o lo que sea.

```python
#Ejemplo 1: Planificador de compras
#Hacer un planificador de compras que se detenga cuando ya no tenga dinero
```

```python
ahorro=int(input("con cuanto dinero cuento?"))
opc=input("Deseo comprar algo? ")
while ahorro>0 and opc=="s":
  precio=int(input("cuanto cuesta eso?"))
  if precio<=ahorro:
    print("Lo compro!!")
    ahorro=ahorro-precio #acumulador
  else:
    print("No me alcanza")

    opc=input("Deseo comprar algo? ")
```
